# 🛢️ Ενότητα 8: Αντικείμενα Βάσης Δεδομένων

Μια βάση δεν αποτελείται μόνο από πίνακες. Περιέχει Όψεις (Views), Ευρετήρια (Indexes), και άλλα αντικείμενα.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Employees (ID INTEGER, Name TEXT, Salary REAL)")
conn.executemany("INSERT INTO Employees VALUES (?, ?, ?)", [(1, 'Alice', 3000), (2, 'Bob', 4000), (3, 'Charlie', 2000)])
conn.commit()
```

## Views (Όψεις)
Μια View είναι σαν ένας εικονικός πίνακας. Ορίζεται από ένα ερώτημα SQL και δεν αποθηκεύει τα δεδομένα ξανά (απλά αποθηκεύει το ερώτημα).

```python
cursor = conn.cursor()
cursor.execute("""
CREATE VIEW HighPaidEmployees AS
SELECT * FROM Employees WHERE Salary > 2500;
""")
conn.commit()

# Τώρα μπορούμε να κάνουμε SELECT κατευθείαν από τη View
pd.read_sql_query("SELECT * FROM HighPaidEmployees;", conn)
```

## Indexes (Ευρετήρια)
Τα Indexes επιταχύνουν δραματικά τις αναζητήσεις σε μεγάλους πίνακες (σαν το ευρετήριο σε ένα βιβλίο). Προσοχή όμως: καθυστερούν τα INSERT/UPDATE/DELETE.

```python
cursor.execute("CREATE INDEX idx_emp_name ON Employees(Name);")
conn.commit()
print("Το ευρετήριο δημιουργήθηκε!")
```
