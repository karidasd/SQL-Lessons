# ✅ Λύσεις: Ενότητα 8

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Products (ID INTEGER, Name TEXT, Price REAL, Category TEXT)")
conn.executemany("INSERT INTO Products VALUES (?, ?, ?, ?)", [(1, 'A', 10, 'Cat1'), (2, 'B', 50, 'Cat1'), (3, 'C', 30, 'Cat2')])
conn.commit()
```

### Λύση 1

```python
conn.execute("CREATE VIEW ExpensiveProducts AS SELECT * FROM Products WHERE Price > 20;")
conn.commit()
```

### Λύση 2

```python
pd.read_sql_query("SELECT * FROM ExpensiveProducts;", conn)
```
