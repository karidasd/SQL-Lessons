# ✅ Λύσεις: Ενότητα 4

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Orders (ID INTEGER, Region TEXT, Sales REAL)")
conn.executemany("INSERT INTO Orders VALUES (?, ?, ?)", [(1, 'North', 500), (2, 'South', 300), (3, 'North', 700), (4, 'East', 400), (5, 'South', 200)])
conn.commit()
```

### Λύση 1

```python
query = "SELECT Region, SUM(Sales) as TotalSales FROM Orders GROUP BY Region;"
pd.read_sql_query(query, conn)
```

### Λύση 2

```python
query = "SELECT Region, SUM(Sales) as TotalSales FROM Orders GROUP BY Region HAVING SUM(Sales) > 600;"
pd.read_sql_query(query, conn)
```
