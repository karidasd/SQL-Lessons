# 🛢️ Ενότητα 4: Συναρτήσεις και Ομαδοποίηση

Συχνά θέλουμε να υπολογίσουμε στατιστικά (άθροισμα, μέσος όρος) για ομάδες δεδομένων. Εδώ έρχονται οι συναρτήσεις συνάθροισης (Aggregation) και το `GROUP BY`.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Sales (ID INTEGER, Category TEXT, Amount REAL)")
conn.executemany("INSERT INTO Sales VALUES (?, ?, ?)", [
    (1, 'A', 100), (2, 'B', 200), (3, 'A', 150), (4, 'B', 50), (5, 'C', 300)
])
conn.commit()
```

## Συναρτήσεις Συνάθροισης
- `COUNT()`: Πλήθος γραμμών
- `SUM()`: Άθροισμα
- `AVG()`: Μέσος όρος
- `MIN()`, `MAX()`: Ελάχιστο, Μέγιστο

```python
query = "SELECT COUNT(*) as TotalSales, SUM(Amount) as TotalAmount, AVG(Amount) as AvgAmount FROM Sales;"
pd.read_sql_query(query, conn)
```

## GROUP BY
Μας επιτρέπει να ομαδοποιήσουμε αποτελέσματα ανά κατηγορία.

```python
query = "SELECT Category, SUM(Amount) as TotalAmount FROM Sales GROUP BY Category;"
pd.read_sql_query(query, conn)
```

## HAVING
Το `HAVING` είναι σαν το `WHERE`, αλλά χρησιμοποιείται για φιλτράρισμα **μετά** το `GROUP BY` (δηλαδή πάνω στα αθροίσματα/στατιστικά).

```python
query = "SELECT Category, SUM(Amount) as TotalAmount FROM Sales GROUP BY Category HAVING TotalAmount > 200;"
pd.read_sql_query(query, conn)
```
