# 🛢️ Ενότητα 6: Προχωρημένα Ερωτήματα

Σε αυτή την ενότητα θα δούμε πώς να γράφουμε πιο πολύπλοκα ερωτήματα, όπως Subqueries (Υποερωτήματα) και CTEs (Common Table Expressions).

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Products (ID INTEGER, Name TEXT, Price REAL)")
conn.executemany("INSERT INTO Products VALUES (?, ?, ?)", [(1, 'A', 10), (2, 'B', 50), (3, 'C', 30)])
conn.commit()
```

## Subqueries (Υποερωτήματα)
Ένα ερώτημα μέσα σε ένα άλλο. Ας βρούμε τα προϊόντα που έχουν τιμή μεγαλύτερη από τον μέσο όρο.

```python
query = """
SELECT Name, Price 
FROM Products 
WHERE Price > (SELECT AVG(Price) FROM Products);
"""
pd.read_sql_query(query, conn)
```

## CTEs (Common Table Expressions - WITH)
Τα CTEs κάνουν τον κώδικά μας πιο ευανάγνωστο σπάζοντας πολύπλοκα subqueries σε λογικά βήματα.

```python
query = """
WITH AvgPrice AS (
    SELECT AVG(Price) as avg_p FROM Products
)
SELECT Name, Price 
FROM Products, AvgPrice 
WHERE Price > AvgPrice.avg_p;
"""
pd.read_sql_query(query, conn)
```

## CASE WHEN
Είναι το αντίστοιχο του if-else στην SQL.

```python
query = """
SELECT Name, Price,
CASE
    WHEN Price > 40 THEN 'Expensive'
    WHEN Price > 20 THEN 'Moderate'
    ELSE 'Cheap'
END as Category
FROM Products;
"""
pd.read_sql_query(query, conn)
```
