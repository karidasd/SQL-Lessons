# ✅ Λύσεις: Ενότητα 6

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Employees (ID INTEGER, Name TEXT, Salary REAL)")
conn.executemany("INSERT INTO Employees VALUES (?, ?, ?)", [(1, 'Alice', 3000), (2, 'Bob', 4000), (3, 'Charlie', 2000), (4, 'David', 5000)])
conn.commit()
```

### Λύση 1

```python
query = "SELECT Name, Salary FROM Employees WHERE Salary = (SELECT MAX(Salary) FROM Employees);"
pd.read_sql_query(query, conn)
```

### Λύση 2

```python
query = "SELECT Name, Salary, CASE WHEN Salary > 3500 THEN 'High' ELSE 'Low' END as Status FROM Employees;"
pd.read_sql_query(query, conn)
```
