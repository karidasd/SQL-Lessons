# 🛢️ Ενότητα 5: Συνενώσεις (Joins)

Μια σχεσιακή βάση δεδομένων μοιράζει τα δεδομένα σε πολλούς πίνακες. Για να τα ενώσουμε και να πάρουμε χρήσιμη πληροφορία, χρησιμοποιούμε τα **JOINS**.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Users (ID INTEGER, Name TEXT)")
conn.execute("CREATE TABLE Orders (OrderID INTEGER, UserID INTEGER, Product TEXT)")

conn.executemany("INSERT INTO Users VALUES (?, ?)", [(1, 'Γιώργος'), (2, 'Μαρία'), (3, 'Κώστας')])
conn.executemany("INSERT INTO Orders VALUES (?, ?, ?)", [(101, 1, 'Laptop'), (102, 1, 'Mouse'), (103, 2, 'Keyboard')])
conn.commit()
```

## INNER JOIN
Επιστρέφει εγγραφές που έχουν αντιστοιχία **ΚΑΙ στους δύο** πίνακες.

```python
query = """
SELECT Users.Name, Orders.Product 
FROM Users
INNER JOIN Orders ON Users.ID = Orders.UserID;
"""
pd.read_sql_query(query, conn)
```

## LEFT JOIN
Επιστρέφει **όλες** τις εγγραφές από τον αριστερό πίνακα, και όσες ταιριάζουν από τον δεξιό. Αν δεν ταιριάζουν, βάζει `NULL`. (Ο Κώστας δεν έχει παραγγελίες)

```python
query = """
SELECT Users.Name, Orders.Product 
FROM Users
LEFT JOIN Orders ON Users.ID = Orders.UserID;
"""
pd.read_sql_query(query, conn)
```
