# ✅ Λύσεις: Ενότητα 5

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Departments (DeptID INTEGER, DeptName TEXT)")
conn.execute("CREATE TABLE Employees (EmpID INTEGER, EmpName TEXT, DeptID INTEGER)")
conn.executemany("INSERT INTO Departments VALUES (?, ?)", [(1, 'IT'), (2, 'HR'), (3, 'Sales')])
conn.executemany("INSERT INTO Employees VALUES (?, ?, ?)", [(1, 'Alice', 1), (2, 'Bob', 1), (3, 'Charlie', 2)])
conn.commit()
```

### Λύση 1

```python
query = "SELECT Employees.EmpName, Departments.DeptName FROM Employees INNER JOIN Departments ON Employees.DeptID = Departments.DeptID;"
pd.read_sql_query(query, conn)
```

### Λύση 2

```python
query = "SELECT Departments.DeptName, Employees.EmpName FROM Departments LEFT JOIN Employees ON Departments.DeptID = Employees.DeptID;"
pd.read_sql_query(query, conn)
```
