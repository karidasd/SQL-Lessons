## Τι είναι η SQL;
Η SQL είναι η γλώσσα που χρησιμοποιούμε για να επικοινωνήσουμε με Σχεσιακές Βάσεις Δεδομένων (Relational Databases). Οι βάσεις αυτές αποθηκεύουν δεδομένα σε **πίνακες** (tables), οι οποίοι αποτελούνται από γραμμές (rows) και στήλες (columns).

## Ρύθμιση Περιβάλλοντος
Στα Jupyter Notebooks, μπορούμε να τρέχουμε SQL χρησιμοποιώντας in-memory βάσεις δεδομένων SQLite. Ας φορτώσουμε τη βιβλιοθήκη `sqlite3` και ας φτιάξουμε ένα προσωρινό πίνακα για να παίξουμε.

````python
import sqlite3
import pandas as pd

# Δημιουργία σύνδεσης με in-memory βάση
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Δημιουργία ενός απλού πίνακα Employees
cursor.execute("""
CREATE TABLE Employees (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary REAL
)
""")

# Εισαγωγή δεδομένων
cursor.executemany("""
INSERT INTO Employees (Name, Department, Salary)
VALUES (?, ?, ?)
""", [
    ("Γιώργος", "IT", 2000),
    ("Μαρία", "HR", 1800),
    ("Κώστας", "IT", 2200),
    ("Ελένη", "Sales", 1900)
])
conn.commit()

print("Η βάση δημιουργήθηκε επιτυχώς!")
````

## Η εντολή SELECT
Η εντολή `SELECT` χρησιμοποιείται για να ανακτήσουμε δεδομένα από έναν πίνακα. Η εντολή `FROM` καθορίζει τον πίνακα.

```sql
SELECT στήλη1, στήλη2 FROM πίνακας;
```

````python
# Χρησιμοποιούμε pandas για να βλέπουμε όμορφα τα αποτελέσματα
query = "SELECT Name, Department FROM Employees;"
df = pd.read_sql_query(query, conn)
df
````

Αν θέλουμε όλες τις στήλες, χρησιμοποιούμε τον αστερίσκο `*`.

````python
query = "SELECT * FROM Employees;"
df = pd.read_sql_query(query, conn)
df
````

## Το φίλτρο WHERE
Για να φιλτράρουμε τις εγγραφές, προσθέτουμε την εντολή `WHERE`.

```sql
SELECT * FROM πίνακας WHERE συνθήκη;
```

````python
query = "SELECT * FROM Employees WHERE Department = 'IT';"
df = pd.read_sql_query(query, conn)
df
````

Μπορούμε να χρησιμοποιήσουμε τελεστές σύγκρισης όπως `>`, `<`, `>=`, `<=`, `!=`. Ας βρούμε όσους έχουν μισθό πάνω από 1900:

````python
query = "SELECT * FROM Employees WHERE Salary > 1900;"
df = pd.read_sql_query(query, conn)
df
````

## Ταξινόμηση με το ORDER BY
Για να ταξινομήσουμε τα αποτελέσματα, χρησιμοποιούμε το `ORDER BY` (προεπιλογή αύξουσα ASC, για φθίνουσα DESC).

````python
query = "SELECT * FROM Employees ORDER BY Salary DESC;"
df = pd.read_sql_query(query, conn)
df
````
