# ✅ Λύσεις: Ενότητα 1
Ακολουθούν οι λύσεις για τις ασκήσεις της Ενότητας 1.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
conn.execute("""
CREATE TABLE Products (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Category TEXT,
    Price REAL
)
""")
conn.executemany("INSERT INTO Products (Name, Category, Price) VALUES (?, ?, ?)", [
    ("Laptop", "Electronics", 1200),
    ("Mouse", "Electronics", 25),
    ("Desk", "Furniture", 150),
    ("Chair", "Furniture", 85),
    ("Headphones", "Electronics", 80)
])
conn.commit()
```

### Λύση 1
Γράψτε ένα ερώτημα SQL που να επιστρέφει το `Name` και την `Price` όλων των προϊόντων.

```python
query = "SELECT Name, Price FROM Products;"
pd.read_sql_query(query, conn)
```

### Λύση 2
Βρείτε όλα τα προϊόντα που ανήκουν στην κατηγορία "Electronics".

```python
query = "SELECT * FROM Products WHERE Category = 'Electronics';"
pd.read_sql_query(query, conn)
```

### Λύση 3
Βρείτε όλα τα προϊόντα με τιμή μεγαλύτερη από 50, ταξινομημένα από το φθηνότερο στο ακριβότερο.

```python
query = "SELECT * FROM Products WHERE Price > 50 ORDER BY Price ASC;"
pd.read_sql_query(query, conn)
```
