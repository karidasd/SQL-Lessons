# 🛢️ Ενότητα 2: Χειρισμός Δεδομένων (DML)

Σε αυτή την ενότητα θα μάθουμε πώς να προσθέτουμε, να ενημερώνουμε και να διαγράφουμε δεδομένα. Αυτές οι λειτουργίες ονομάζονται **DML** (Data Manipulation Language).

````python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE Users (
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT,
    Age INTEGER,
    City TEXT
)
""")
conn.commit()
print("Ο πίνακας Users είναι έτοιμος!")
````

## INSERT (Εισαγωγή)
Η εντολή `INSERT INTO` χρησιμοποιείται για να προσθέσουμε νέες γραμμές στον πίνακα.

```sql
INSERT INTO πίνακας (στήλη1, στήλη2) VALUES (τιμή1, τιμή2);
```

````python
cursor.execute("INSERT INTO Users (Name, Age, City) VALUES ('Νίκος', 28, 'Αθήνα');")
cursor.execute("INSERT INTO Users (Name, Age, City) VALUES ('Μαρία', 24, 'Θεσσαλονίκη');")
cursor.execute("INSERT INTO Users (Name, Age, City) VALUES ('Πέτρος', 35, 'Πάτρα');")
conn.commit()

pd.read_sql_query("SELECT * FROM Users;", conn)
````

Μπορούμε να κάνουμε και πολλαπλές εισαγωγές (multi-row insert):

````python
cursor.execute("""
INSERT INTO Users (Name, Age, City)
VALUES 
    ('Ελένη', 30, 'Ηράκλειο'),
    ('Γιάννης', 22, 'Αθήνα');
""")
conn.commit()
pd.read_sql_query("SELECT * FROM Users;", conn)
````

## UPDATE (Ενημέρωση)
Η εντολή `UPDATE` χρησιμοποιείται για να αλλάξουμε υπάρχοντα δεδομένα. **Προσοχή!** Πάντα να χρησιμοποιείτε `WHERE`, διαφορετικά θα ενημερωθούν όλες οι γραμμές!

```sql
UPDATE πίνακας SET στήλη = νέα_τιμή WHERE συνθήκη;
```

````python
# Ας αλλάξουμε την ηλικία του Νίκου
cursor.execute("UPDATE Users SET Age = 29 WHERE Name = 'Νίκος';")
conn.commit()
pd.read_sql_query("SELECT * FROM Users;", conn)
````

## DELETE (Διαγραφή)
Η εντολή `DELETE FROM` διαγράφει γραμμές. Όπως και με το UPDATE, μην ξεχνάτε το `WHERE`!

```sql
DELETE FROM πίνακας WHERE συνθήκη;
```

````python
# Ας διαγράψουμε όσους μένουν στην Πάτρα
cursor.execute("DELETE FROM Users WHERE City = 'Πάτρα';")
conn.commit()
pd.read_sql_query("SELECT * FROM Users;", conn)
````
