# ✅ Λύσεις: Ενότητα 2

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Inventory (ID INTEGER PRIMARY KEY AUTOINCREMENT, Item TEXT, Quantity INTEGER)")
conn.executemany("INSERT INTO Inventory (Item, Quantity) VALUES (?, ?)", [("Apples", 50), ("Bananas", 30), ("Oranges", 20)])
conn.commit()
```

### Λύση 1

```python
conn.execute("INSERT INTO Inventory (Item, Quantity) VALUES ('Grapes', 100);")
conn.commit()
pd.read_sql_query("SELECT * FROM Inventory;", conn)
```

### Λύση 2

```python
conn.execute("UPDATE Inventory SET Quantity = 40 WHERE Item = 'Bananas';")
conn.commit()
pd.read_sql_query("SELECT * FROM Inventory;", conn)
```

### Λύση 3

```python
conn.execute("DELETE FROM Inventory WHERE Item = 'Oranges';")
conn.commit()
pd.read_sql_query("SELECT * FROM Inventory;", conn)
```
