# 🛢️ Ενότητα 7: Συναρτήσεις Παραθύρου (Window Functions)

Οι συναρτήσεις παραθύρου εκτελούν υπολογισμούς σε ένα σύνολο γραμμών (ένα "παράθυρο" δεδομένων) που σχετίζονται με την τρέχουσα γραμμή. Σε αντίθεση με το `GROUP BY`, δεν "συμπιέζουν" τις γραμμές.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Employees (Name TEXT, Dept TEXT, Salary REAL)")
conn.executemany("INSERT INTO Employees VALUES (?, ?, ?)", [
    ('Alice', 'IT', 3000), ('Bob', 'IT', 4000), ('Charlie', 'IT', 3500),
    ('David', 'HR', 2000), ('Eve', 'HR', 2500)
])
conn.commit()
```

## OVER() και PARTITION BY
Χρησιμοποιούνται για να ορίσουν το παράθυρο. Ας βρούμε τον μέσο μισθό ανά τμήμα, διατηρώντας όμως όλες τις γραμμές.

```python
query = """
SELECT Name, Dept, Salary,
       AVG(Salary) OVER (PARTITION BY Dept) as AvgDeptSalary
FROM Employees;
"""
pd.read_sql_query(query, conn)
```

## RANK() και ROW_NUMBER()
Χρησιμοποιούνται για να βαθμολογήσουμε ή να αριθμήσουμε γραμμές μέσα σε ένα παράθυρο.

```python
query = """
SELECT Name, Dept, Salary,
       RANK() OVER (PARTITION BY Dept ORDER BY Salary DESC) as RankInDept,
       ROW_NUMBER() OVER (PARTITION BY Dept ORDER BY Salary DESC) as RowNumInDept
FROM Employees;
"""
pd.read_sql_query(query, conn)
```
