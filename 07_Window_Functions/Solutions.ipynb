```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
conn.execute("CREATE TABLE Sales (Salesperson TEXT, Region TEXT, Amount REAL)")
conn.executemany("INSERT INTO Sales VALUES (?, ?, ?)", [('John', 'North', 500), ('Mike', 'North', 600), ('Anna', 'North', 500), ('Chris', 'South', 400), ('Sara', 'South', 700)])
conn.commit()
```

### Λύση 1

```python
query = """
SELECT Salesperson, Region, Amount,
       RANK() OVER (PARTITION BY Region ORDER BY Amount DESC) as SalesRank
FROM Sales;
"""
pd.read_sql_query(query, conn)
```
