# ✅ Λύσεις: Ενότητα 3

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
```

### Λύση 1

```python
conn.execute("""
CREATE TABLE Books (
    BookID INTEGER PRIMARY KEY,
    Title TEXT NOT NULL,
    Author TEXT,
    Price REAL
);
""")
conn.commit()
```

### Λύση 2

```python
conn.execute("ALTER TABLE Books ADD COLUMN PublishedYear INTEGER;")
conn.commit()
pd.read_sql_query("PRAGMA table_info(Books);", conn)
```
