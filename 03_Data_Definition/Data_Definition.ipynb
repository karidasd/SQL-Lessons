# 🛢️ Ενότητα 3: Ορισμός Δεδομένων (DDL)

Η DDL (Data Definition Language) χρησιμοποιείται για να ορίζουμε, να τροποποιούμε και να διαγράφουμε τη δομή των πινάκων στη βάση δεδομένων.

## CREATE TABLE
Χρησιμοποιείται για τη δημιουργία νέων πινάκων. Καθορίζουμε τα ονόματα των στηλών και τους τύπους δεδομένων (π.χ. `INTEGER`, `TEXT`, `REAL`). Μπορούμε επίσης να προσθέσουμε **Constraints** (περιορισμούς) όπως `PRIMARY KEY` ή `NOT NULL`.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE Students (
    StudentID INTEGER PRIMARY KEY,
    FirstName TEXT NOT NULL,
    LastName TEXT NOT NULL,
    EnrollmentYear INTEGER DEFAULT 2024
);
""")
conn.commit()
print("Ο πίνακας δημιουργήθηκε!")
```

## ALTER TABLE
Χρησιμοποιείται για να αλλάξουμε τη δομή ενός υπάρχοντος πίνακα (π.χ. να προσθέσουμε μια στήλη).

```python
cursor.execute("ALTER TABLE Students ADD COLUMN Email TEXT;")
conn.commit()

# Ας δούμε τη δομή του πίνακα στο SQLite
pd.read_sql_query("PRAGMA table_info(Students);", conn)
```

## DROP TABLE
Διαγράφει **πλήρως** έναν πίνακα και όλα τα δεδομένα του. ΠΡΟΣΟΧΗ!

```python
cursor.execute("DROP TABLE Students;")
conn.commit()
print("Ο πίνακας διαγράφηκε.")
```
